## (1) Derivation of wave equation
The dispersion relation of electromagnetic waves can be found from Maxwell's equations:
\begin{equation} 
\nabla\times \mathbf{E} = -\frac{1}{c}\frac{\partial\mathbf{B}}{\partial t} \quad\quad \text{(Faraday)},\label{eq:Faraday}
\end{equation}

\begin{equation}
\nabla\times\mathbf{B} = \frac{1}{c}\left(\frac{\partial\mathbf{E}}{\partial t} + 4\pi\mathbf{J}\right)=\frac{1}{c}\left(\frac{\partial\mathbf{E}}{\partial t} + 4\pi\overleftrightarrow{\sigma}\cdot\mathbf{E}\right) \quad\quad\text{(Amp√®re)}, \label{eq:Ampere}
\end{equation}

by eliminating the perturbed magnetic field $\mathbf{B}$ (or electric field $\mathbf{E}$) between these equations. It is convenient to introduce a dielectric tensor defined by

\begin{equation}
\mathbf{K} = 1 + i\frac{4\pi}{\omega}\overleftrightarrow{\sigma}.
\end{equation}

Eliminating $\mathbf{B}$ between Eqs. (\ref{eq:Faraday}) and (\ref{eq:Ampere}), we obtain the wave equation (note that $\nabla\times\nabla\times\mathbf{A}=\nabla(\nabla\cdot\mathbf{A})-\nabla^2\mathbf{A}$):

\begin{equation}
\nabla(\nabla\cdot\mathbf{E}) - \nabla^2\mathbf{E}=\frac{\omega^2}{c^2}\mathbf{K}\cdot\mathbf{E}, \label{eq:wave}
\end{equation}

The general solution of a system of linear equations can be constructed as a superposition of plane wave solutions:

\begin{equation}
\mathbf{E}(\mathbf{r},t)=\mathbf{E}_k\exp[i(\mathbf{k}\cdot r-\omega t)] ,
\end{equation}

therefore, Eq.(\ref{eq:wave}) can be written as:

\begin{equation}
\mathbf{M}\cdot\mathbf{E}=0,
\end{equation}

where

\begin{equation}
\mathbf{M}=\mathbf{kk}-k^2\mathbf{I}+\frac{\omega^2}{c^2}\mathbf{K}.
\end{equation}

The condition

\begin{equation}
M(\omega,\mathbf{k})\equiv\text{det}(\mathbf{M}) = 0,
\end{equation}

is called the dispersion relation.
## (2) Derivation of the cold plasma dielectric tensor
Linearized momentum equation is:

\begin{equation}
m_sn_s\frac{\partial\mathbf{V_s}}{\partial t} = e_sn_s(\mathbf{E} + \mathbf{V_s}\times\mathbf{B_0}).
\end{equation}

Using the definition $\mathbf{j}=\sum_{s}e_sn_s\mathbf{V}_s$, we can obtain

\begin{equation}
\mathbf{K} = \begin{pmatrix}
  \text{S} & \text{iD} & 0 \\
  \text{-iD} & \text{S} & 0 \\
  0 & 0 & P
 \end{pmatrix},
\end{equation}

where 

\begin{equation}
\text{S} = \frac{\text{R+L}}{2},
\end{equation}

and

\begin{equation}
\text{D} = \frac{\text{R-L}}{2}.
\end{equation}

The exact expressions for $\text{R}$, $\text{L}$ and $\text{P}$, which are most easily obtained by solving the individual charged particle equations of motion, and then summing to obtain the fluid response, are: 

\begin{equation}
\text{R} = 1 - \frac{\omega_{pe}^2}{\omega^2}\left(\frac{\omega}{\omega + \Omega_e}\right) - \frac{\omega_{pi}^2}{\omega^2}\left(\frac{\omega}{\omega + \Omega_i}\right),
\end{equation}

\begin{equation}
\text{L} = 1 - \frac{\omega_{pe}^2}{\omega^2}\left(\frac{\omega}{\omega - \Omega_e}\right) - \frac{\omega_{pi}^2}{\omega^2}\left(\frac{\omega}{\omega - \Omega_i}\right),
\end{equation}

\begin{equation}
\text{P} = 1 - \frac{\omega_{pe}^2}{\omega^2} - \frac{\omega_{pi}^2}{\omega^2}.
\end{equation}

## (3) Dispersion relation
let us assume $E_z = B_x = B_y = 0$ and $k = k_x$, then the eigenmode equation simplifies to

\begin{equation}
\begin{pmatrix}
  \text{S} & \text{iD}\\
  \text{-iD} & \text{S}-k^2
   \end{pmatrix}\begin{pmatrix}
  E_x\\
  E_y
   \end{pmatrix} = 0,
\end{equation}

this dispersion relation reduces to

\begin{equation}
k^2 = \frac{RL}{S}
\end{equation}
